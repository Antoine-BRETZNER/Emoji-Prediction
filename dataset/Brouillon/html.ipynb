```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

# Chemin vers votre chromedriver
PATH = '/home/antoine/Téléchargements/chromedriver-linux64/chromedriver'

# Configuration du driver Chrome
chrome_options = Options()
chrome_options.add_argument("--headless")
service = Service(PATH)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Ouvrir la page Twitter
target_url = "https://x.com/search?q=%23emoji&src=typed_query&f=live"
driver.get(target_url)
time.sleep(5)

# Faire défiler la page pour charger plus de contenu
scroll_pause_time = 2  # Temps d'attente après chaque défilement

# Obtenez la hauteur initiale de la page
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Faites défiler vers le bas de la page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Attendez que la page charge plus de contenu
    time.sleep(scroll_pause_time)

    # Calculez la nouvelle hauteur de la page et comparez-la avec la hauteur précédente
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break  # Arrêtez le défilement lorsque la fin de la page est atteinte
    last_height = new_height

# Obtenir le contenu de la page après le chargement
resp = driver.page_source
driver.quit()

# Analyser le contenu HTML avec BeautifulSoup
soup = BeautifulSoup(resp, 'html.parser')

# Extraire toutes les divs ayant data-testid="tweetText"
divs = soup.find_all('div', {'data-testid': 'tweetText'})

# Extraire et imprimer le texte de chaque div, y compris les emojis
tweets = []
for div in divs:
    tweet_parts = []
    for element in div.children:
        if element.name == 'img':
            tweet_parts.append(element['alt'])  # Ajouter l'emoji
        else:
            tweet_parts.append(element.get_text())  # Ajouter le texte
    tweet_text = ''.join(tweet_parts)
    tweets.append(tweet_text)

for tweet in tweets:
    print(tweet)
    print("-" * 80)  # Séparateur pour chaque tweet
```
